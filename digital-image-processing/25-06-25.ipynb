# Read and display image

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread("./images/cat-1.jpg") # read an image (color image)
cv2.imshow("Original", img) # display the image
cv2.waitKey(0) # wait before the image display window before closing
```

# Convert RGB to Grayscale image

```python
img2 = cv2.imread("./images/cat-1.jpg")
rgbToGrayScale = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
cv2.imshow("Grayscale image", rgbToGrayScale)
cv2.waitKey(0)
```

# Blur image

```python
img3 = cv2.imread("./images/cat-2.jpg")
blurImg = cv2.GaussianBlur(img3, (11, 11), 0)
cv2.imshow("Gaussian Blur image", blurImg)
cv2.waitKey(0)
```

# Resize image

```python
img4 = cv2.imread("./images/cat-2.jpg")
resized_img = cv2.resize(img4, (300, 300))
cv2.imshow("Resized image", resized_img)
cv2.waitKey(0)
```

# Remove noise from image

```python
img6 = cv2.imread("./images/noise-img.jpg") # image with noise

# apply gaussian blur to reduce noise
blurred = cv2.GaussianBlur(img6, (11, 11), 0)
cv2.imshow("Blurred image", blurred)
cv2.waitKey(0)
```
